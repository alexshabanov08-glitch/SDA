```python
%%writefile lab1.c

#include <stdio.h>
#include <string.h>

#define MAX_SIZE 10

// Структура данных
struct Request {
    int number;           // Номер заявки
    char description[100];
    char status[20];
    char date[20];
};

// Функции
void addRequest(struct Request arr[], int *count);
void printRequests(struct Request arr[], int count);
void editRequest(struct Request arr[], int count);
void searchRequest(struct Request arr[], int count);

int main() {
    struct Request requests[MAX_SIZE];
    int count = 0;
    int choice;

    do {
        printf("\nСИСТЕМА УЧЕТА ЗАЯВОК (Записей: %d из %d)\n", count, MAX_SIZE);
        printf("1 - Добавить новую заявку\n");
        printf("2 - Вывести список всех заявок\n");
        printf("3 - Редактировать заявку по ID\n");
        printf("4 - Поиск по номеру заявки\n");
        printf("5 - Выход\n");
        printf("Выберите действие: ");
        fflush(stdout);

        if (scanf("%d", &choice) != 1) break;

        switch(choice) {
            case 1:
                addRequest(requests, &count);
                break;
            case 2:
                printRequests(requests, count);
                break;
            case 3:
                editRequest(requests, count);
                break;
            case 4:
                searchRequest(requests, count);
                break;
        }
    } while(choice != 5);

    return 0;
}

// Добавление заявки
void addRequest(struct Request arr[], int *count) {
    if (*count >= MAX_SIZE) {
        printf("\nОшибка: Память заполнена (максимум %d заявок)\n", MAX_SIZE);
        return;
    }
    int i = *count;
    printf("\n--- Ввод данных для заявки под ID %d ---\n", i + 1);

    printf("Введите номер заявки: ");
    fflush(stdout);
    scanf("%d", &arr[i].number);

    printf("Описание проблемы: ");
    fflush(stdout);
    scanf(" %[^\n]", arr[i].description);

    printf("Статус (в обработке/завершено): ");
    fflush(stdout);
    scanf(" %[^\n]", arr[i].status);

    printf("Дата подачи (ДД.ММ.ГГ): ");
    fflush(stdout);
    scanf(" %[^\n]", arr[i].date);

    (*count)++;
    printf("Заявка успешно добавлена!\n");
}

// Вывод всех заявок
void printRequests(struct Request arr[], int count) {
    if (count == 0) {
        printf("\nСписок заявок пуст.\n");
        return;
    }
    // Шапка таблицы (добавили колонку Дата)
    printf("\n--------------------------------------------------------------------------------\n");
    printf("%-4s | %-10s | %-20s | %-15s | %-12s\n", "ID", "№ Заявки", "Описание", "Статус", "Дата");
    printf("--------------------------------------------------------------------------------\n");

    for(int i = 0; i < count; i++) {

        printf("%-4d | %-10d | %-20s | %-15s | %-12s\n",
                i + 1,
                arr[i].number,
                arr[i].description,
                arr[i].status,
                arr[i].date);
    }
}

// Редактирование заявки по ID (позиции в списке)
void editRequest(struct Request arr[], int count) {
    if (count == 0) {
        printf("\nСписок пуст, нечего редактировать.\n");
        return;
    }

    int displayID, choice;
    printf("Введите ID заявки из списка для изменения (1 - %d): ", count);
    fflush(stdout);
    scanf("%d", &displayID);

    int idx = displayID - 1; // превращение ID в индекс массива

    if(idx < 0 || idx >= count) {
        printf("Ошибка: Заявка с таким ID не существует.\n");
        return;
    }

    printf("\nЧто изменить в заявке ID %d?\n", displayID);
    printf("1 - Номер заявки\n2 - Описание\n3 - Статус\n4 - Дату\nВыбор: ");
    fflush(stdout);
    scanf("%d", &choice);

    switch(choice) {
        case 1:
            printf("Новый номер: "); fflush(stdout);
            scanf("%d", &arr[idx].number); break;
        case 2:
            printf("Новое описание: "); fflush(stdout);
            scanf(" %[^\n]", arr[idx].description); break;
        case 3:
            printf("Новый статус: "); fflush(stdout);
            scanf(" %[^\n]", arr[idx].status); break;
        case 4:
            printf("Новая дата: "); fflush(stdout);
            scanf(" %[^\n]", arr[idx].date); break;
        default:
            printf("Действие отменено.\n");
    }
}

// Поиск заявки по номеру
void searchRequest(struct Request arr[], int count) {
    int searchNum;
    printf("Введите номер заявки для поиска: ");
    fflush(stdout);
    scanf("%d", &searchNum);

    for(int i = 0; i < count; i++) {
        if(arr[i].number == searchNum) {
            printf("\n[Найдено] ID в списке: %d\n", i + 1);
            printf("Описание: %s\nСтатус: %s\nДата: %s\n",
                    arr[i].description, arr[i].status, arr[i].date);
            return;
        }
    }
    printf("Заявка с номером %d не найдена.\n", searchNum);
}
```
