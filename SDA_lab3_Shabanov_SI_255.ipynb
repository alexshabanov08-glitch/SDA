```python
%%writefile lab3.c

#include <stdio.h>
#include <string.h>
#include <stdlib.h>

#define INITIAL_SIZE 2

// Структура данных
struct Request {
    int number;
    char description[100];
    char status[20];
    char date[20];
};

// Функции
void addRequest(struct Request **arr, int *count, int *capacity);
void printRequests(struct Request arr[], int count);
void editRequest(struct Request arr[], int count);
void searchRequest(struct Request arr[], int count);
void expandArray(struct Request **arr, int count, int *capacity, int extra);
void removeLast(struct Request **arr, int *count, int *capacity);
void clearMemory(struct Request **arr);

void sortRequests(struct Request arr[], int count);
void insertRequest(struct Request **arr, int *count, int *capacity);
void deleteByIndex(struct Request **arr, int *count);

int main() {

    int capacity = INITIAL_SIZE;
    int count = 0;
    int choice;

    struct Request *requests = malloc(capacity * sizeof(struct Request));

    do {
        printf("\nСИСТЕМА УЧЕТА ЗАЯВОК (Записей: %d)\n", count);
        printf("1 - Добавить новую заявку\n");
        printf("2 - Вывести список всех заявок\n");
        printf("3 - Редактировать заявку по ID\n");
        printf("4 - Поиск по номеру заявки\n");
        printf("5 - Удалить последнюю заявку\n");
        printf("6 - Расширить массив вручную\n");
        printf("7 - Очистить память\n");
        printf("8 - Сортировка заявок\n");
        printf("9 - Вставить заявку на позицию\n");
        printf("10 - Удалить заявку по индексу\n");
        printf("11 - Выход\n");

        printf("Выберите действие: ");
        fflush(stdout);

        if (scanf("%d", &choice) != 1) break;

        switch(choice) {

            case 1:
                addRequest(&requests, &count, &capacity);
                break;

            case 2:
                printRequests(requests, count);
                break;

            case 3:
                editRequest(requests, count);
                break;

            case 4:
                searchRequest(requests, count);
                break;

            case 5:
                removeLast(&requests, &count, &capacity);
                break;

            case 6: {
                int extra;
                printf("На сколько элементов расширить массив: ");
                scanf("%d", &extra);
                expandArray(&requests, count, &capacity, extra);
                break;
            }

            case 7:
                clearMemory(&requests);
                count = 0;
                capacity = 0;
                break;

            case 8:
                sortRequests(requests, count);
                break;

            case 9:
                insertRequest(&requests, &count, &capacity);
                break;

            case 10:
                deleteByIndex(&requests, &count);
                break;

        }

    } while(choice != 11);

    clearMemory(&requests);

    return 0;
}

// Добавление заявки
void addRequest(struct Request **arr, int *count, int *capacity) {

    if (*count >= *capacity) {
        expandArray(arr, *count, capacity, 2);
    }

    int i = *count;

    printf("\n--- Ввод данных для заявки под ID %d ---\n", i + 1);

    printf("Введите номер заявки: ");
    scanf("%d", &(*arr)[i].number);

    printf("Описание проблемы: ");
    scanf(" %[^\n]", (*arr)[i].description);

    printf("Статус: ");
    scanf(" %[^\n]", (*arr)[i].status);

    printf("Дата подачи: ");
    scanf(" %[^\n]", (*arr)[i].date);

    (*count)++;

    printf("Заявка успешно добавлена!\n");
}

// Расширение массива
void expandArray(struct Request **arr, int count, int *capacity, int extra) {

    int newCapacity = *capacity + extra;

    struct Request *newArr = malloc(newCapacity * sizeof(struct Request));

    for(int i = 0; i < count; i++) {
        newArr[i] = (*arr)[i];
    }

    free(*arr);

    *arr = newArr;
    *capacity = newCapacity;

    printf("Массив расширен. Новая емкость: %d\n", *capacity);
}

// Удаление последнего
void removeLast(struct Request **arr, int *count, int *capacity) {

    if(*count == 0) {
        printf("Список пуст.\n");
        return;
    }

    (*count)--;

    struct Request *newArr = malloc((*count) * sizeof(struct Request));

    for(int i = 0; i < *count; i++) {
        newArr[i] = (*arr)[i];
    }

    free(*arr);

    *arr = newArr;
    *capacity = *count;

    printf("Последняя заявка удалена.\n");
}

// Очистка памяти
void clearMemory(struct Request **arr) {

    if(*arr != NULL) {
        free(*arr);
        *arr = NULL;
        printf("Память освобождена.\n");
    }
}

// Вывод всех заявок
void printRequests(struct Request arr[], int count) {

    if (count == 0) {
        printf("\nСписок заявок пуст.\n");
        return;
    }

    printf("\n--------------------------------------------------------------------------\n");
    printf("%-4s | %-10s | %-20s | %-15s | %-12s\n", "ID", "№", "Описание", "Статус", "Дата");
    printf("--------------------------------------------------------------------------\n");

    for(int i = 0; i < count; i++) {

        printf("%-4d | %-10d | %-20s | %-15s | %-12s\n",
                i + 1,
                arr[i].number,
                arr[i].description,
                arr[i].status,
                arr[i].date);
    }
}

// Редактирование
void editRequest(struct Request arr[], int count) {

    if (count == 0) {
        printf("\nСписок пуст.\n");
        return;
    }

    int displayID, choice;

    printf("Введите ID заявки (1 - %d): ", count);
    scanf("%d", &displayID);

    int idx = displayID - 1;

    if(idx < 0 || idx >= count) {
        printf("Ошибка: ID не существует.\n");
        return;
    }

    printf("\n1 - Номер\n2 - Описание\n3 - Статус\n4 - Дата\nВыбор: ");
    scanf("%d", &choice);

    switch(choice) {

        case 1:
            scanf("%d", &arr[idx].number);
            break;

        case 2:
            scanf(" %[^\n]", arr[idx].description);
            break;

        case 3:
            scanf(" %[^\n]", arr[idx].status);
            break;

        case 4:
            scanf(" %[^\n]", arr[idx].date);
            break;
    }
}

// Поиск
void searchRequest(struct Request arr[], int count) {

    int searchNum;

    printf("Введите номер заявки: ");
    scanf("%d", &searchNum);

    for(int i = 0; i < count; i++) {

        if(arr[i].number == searchNum) {

            printf("\n[Найдено] ID: %d\n", i + 1);
            printf("Описание: %s\n", arr[i].description);
            printf("Статус: %s\n", arr[i].status);
            printf("Дата: %s\n", arr[i].date);

            return;
        }
    }

    printf("Заявка не найдена.\n");
}

// Сортировка
void sortRequests(struct Request arr[], int count) {

    int choice;

    printf("Сортировать по:\n");
    printf("1 - Номер заявки\n");
    printf("2 - Дата\n");
    scanf("%d", &choice);

    for(int i=0;i<count-1;i++) {
        for(int j=0;j<count-i-1;j++) {

            int swap = 0;

            if(choice==1 && arr[j].number > arr[j+1].number)
                swap=1;

            if(choice==2 && strcmp(arr[j].date,arr[j+1].date)>0)
                swap=1;

            if(swap){
                struct Request temp=arr[j];
                arr[j]=arr[j+1];
                arr[j+1]=temp;
            }
        }
    }

    printf("Сортировка выполнена.\n");
}

// Вставка элемента
void insertRequest(struct Request **arr, int *count, int *capacity) {

    int pos;

    printf("Введите позицию (1 - %d): ", *count + 1);
    scanf("%d", &pos);

    if(pos < 1 || pos > *count + 1){
        printf("Ошибка позиции.\n");
        return;
    }

    if(*count >= *capacity)
        expandArray(arr,*count,capacity,2);

    int idx=pos-1;

    for(int i=*count;i>idx;i--)
        (*arr)[i]=(*arr)[i-1];

    printf("Введите номер: ");
    scanf("%d",&(*arr)[idx].number);

    printf("Описание: ");
    scanf(" %[^\n]",(*arr)[idx].description);

    printf("Статус: ");
    scanf(" %[^\n]",(*arr)[idx].status);

    printf("Дата: ");
    scanf(" %[^\n]",(*arr)[idx].date);

    (*count)++;

    printf("Заявка вставлена.\n");
}

// Удаление по индексу
void deleteByIndex(struct Request **arr, int *count){

    if(*count==0){
        printf("Список пуст.\n");
        return;
    }

    int pos;

    printf("Введите ID (1-%d): ",*count);
    scanf("%d",&pos);

    if(pos<1 || pos>*count){
        printf("Ошибка индекса.\n");
        return;
    }

    int idx=pos-1;

    for(int i=idx;i<*count-1;i++)
        (*arr)[i]=(*arr)[i+1];

    (*count)--;

    printf("Заявка удалена.\n");
}
```
